```python
# there are different activation functions available
# they are
# identy
# binary step
# logistic or sigmoid
# tanh - tan hyperbolic
# arc tan - tan inverse
# rectified linear unit ReLU
# leaky ReLU
# softmax - takes probablity into account
```

```python

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# identy function f(x) = x
x=np.linspace(-10,10,100)
yi=x
plt.plot(x,yi)
# input = output
```

```python
# bisnary step function 
# f(x)= 0 for x<0 and 1 for x>=0
def bi(x):
    if(x<0):
        return 0
    else:
        return 1
y=[]
for i in x:
    y.append(bi(i))
plt.plot(x,y)
# this is good for binary classification 
```

```python
# logistic or sigmoid
# f(x)= 1/(1+e^-x)
yl=1/(1+np.exp(-x))
plt.plot(x,yl)
# this function will map it between 0 and 1
# this is veery useful in neural network
# when the value of x increases exponentially
```

```python
# tan h acticvation functiion
# f(x)= (2/(1+e^-2x))-1
yt=(2/(1+np.exp(-2*x)))-1
plt.plot(x,yt)
# this contains the input between -1 and 1
```

```python
# arc tan function
# f(x)= tan-1(x)
yat=np.arctan(x)
plt.plot(x,yat)
# the maping will be between -pi/2 to pi/2
```

```python
# relu which is also known as rectified linear unit function
# f(x)=0 for x<0 and x for x>=0
def relu(x):
    if(x<0):
        return 0
    else:
        return x
yr=[]
for i in x:
    yr.append(relu(i))
plt.plot(x,yr)
# this function will fundamentally remove the negetive part of the function
```

```python
# leaky relu which is also known as rectified linear unit function
# f(x)=0.01x for x<0 and x for x>=0
def lrelu(x):
    if(x<0):
        return 0.1*x
    else:
        return x
ylr=[]
for i in x:
    ylr.append(lrelu(i))
plt.plot(x,ylr)
# it should be 0.01 
# i have just used 0.1 for visualisation purposes only
# so that we can actually see the gradient
# this function will not make the negetive input to zero just reduces it in magnitude
```

```python
# softmax classifiers this function is used to immpart probablities
# sig(zj)=exp(zj)/(summation k=1 to k=K exp(zk)) for j=1,...K
# this is done lets say u have 5 or 5 outputs then u will get the probablity distribution of each
# this is useful for finding out the most probable ooccurance i.e the classification where the probablity of occorance of a class is maximum
```
